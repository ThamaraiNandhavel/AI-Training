```python
def mean_var(x,y):
    e=sum(v*p for v,p in zip(x,y))
        
    e1=sum(v**2*p for v,p in zip(x,y))
        
    var=e1-e**2
    return e,var
x=[0,1,2]
p_x=[0.3,0.5,0.2]
E,V=mean_var(x,p_x)
print(E)
print(V)
```

```python
#central limit theorm 
#simulation parameter
import matplotlib.pyplot as plt
n_trial=1000
n_dice=30
sample=[]
for i in range(n_trial):
    r=np.random.randint(1,7,n_dice)
    sample.append(np.mean(r))
plt.hist(sample,bins=30,color="red",edgecolor="black",density=True)  
plt.title('Central Limit Theorem Simulation')
plt.xlabel('Sample Mean')   
plt.ylabel('Density')
plt.grid(True)
plt.axvline(x=3.5, color="blue", linestyle="--", label="expected mean(3.5)")
plt.legend()
plt.show()
```
